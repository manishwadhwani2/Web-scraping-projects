Step -1 : Get the Html

The next step is to parse the html content and give it a  tree like structure so it can be easily traversed.



```python
import requests
```

```python
import csv
```

```python
from bs4 import BeautifulSoup
```

```python
!pip install lxml
```

```python
url = "https://techyaapa37.blogspot.com"
```

```python
#Step -1 Get the HTML
r = requests.get(url)
htmlcontent = r.content
```

```python
#step-2 Parse the html
soup = BeautifulSoup(htmlcontent,"lxml")
print(soup.prettify)
```

Step -3 Html tree parser


```python
# get the title of the html page
title  = soup.title.text
print(title)
```

```python
# get all the paragraph of the html page
paragraph = soup.find_all('p')
print(paragraph)
```

```python
# get all the anchor of the html page
anchors = soup.find_all('a')
print(anchors)
```

```python
# print first element of the html page
print(soup.find('p'))
```

```python
# get the class elements of the first element of the html page
print(soup.find("p")["class"])
```

```python
#find all the elements with class lead 
print(soup.find_all("p",class_="description"))
```

```python
# get the all text from the soup
print(soup.get_text())
```

```python
#get all the link from the soup
all_links = set()
anchors = soup.find_all('a')
for link in anchors:
       if(link.get("href")!="#"):
        links = link.get('href')
        all_links.add(links)
        print(links)
        
        
```

```python
navbarsupportedcontent = soup.find(id = 'summary7498192174223721541')
```

```python
for elements in navbarsupportedcontent.children:
    print(elements)
```

```python
for elements in navbarsupportedcontent.strings:
    print(elements)
```

```python
for elements in navbarsupportedcontent.stripped_strings:
    print(elements)
```

```python
for item in navbarsupportedcontent.parent:
    print(item.name)
```

```python
# extracting only text data from the html page using regular expresssion 
import re
data = soup.findAll(text=True)
def visible(element):
    if element.parent.name in ['style', 'script', '[document]', 'head', 'title']:
        return False
    elif re.match('<!--.*-->', str(element.encode('utf-8'))):
        return False
    return True
result = filter(visible, data)
r=list(result)
for k in r:
    if k=="\n":
        r.remove("\n")
print(r)
```

```python
#extracting text data using tags
list1 = []
match = soup.find_all("article");
for element in match:
    print(element.text)
    list1.append(element.text)
```

```python
for k in list1:
    if k=="\n":
        list1.remove("\n")
print(list1[0])
```

```python
#Storing all the title of the articles in the list 
mylist_2 = []
match = soup.find_all("div","block-image")
for element in match:
    print(element.a.img["alt"])
    mylist_2.append(element.a.img["alt"])
```

```python
print(mylist_2[0])
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(list(zip(mylist_2,list1)),columns =["titles","description"])
```
